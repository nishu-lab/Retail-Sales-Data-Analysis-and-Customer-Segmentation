# Retail Sales Data Analysis & Customer Segmentation

## Level 2: Data Cleaning & Preparation

### Objective

The objective of this phase is to improve data quality and prepare the dataset for analysis.

This includes identifying and resolving missing values, duplicate records, invalid transactions, and data consistency issues.

By the end of this phase, the dataset will be cleaned, validated, and ready for feature engineering, exploratory analysis, and customer segmentation.

---

### Environment Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Environment Ready")
```

All required libraries were successfully imported and the analysis environment is ready.

---

### Dataset Loading

```python
df = pd.read_csv("../data/Online Retail.csv", encoding="latin1")
```

The dataset was loaded successfully and is ready for the data cleaning process.

---

### Date Format Conversion

```python
df["InvoiceDate"] = pd.to_datetime(
    df["InvoiceDate"],
    format="mixed"
)
```

The InvoiceDate column was converted to datetime format to support time-based analysis and feature engineering.

---

### Duplicate Records Assessment

```python
df.duplicated().sum()
```

A total of 5,268 duplicate records were identified within the dataset. These records could lead to inflated sales figures and inaccurate analytical results if not addressed.

---

### Duplicate Records Removal


```python
df = df.drop_duplicates()
```

Duplicate transaction records were successfully removed to ensure that each transaction is represented only once within the dataset.

---

### Dataset Dimensions After Duplicate Removal

```python
df.shape
```

After removing duplicate records, the dataset contains 536,641 rows and 8 columns, indicating a reduction in redundant transaction entries.

---

### Missing Values Assessment

```python
df.isnull().sum()
```

Missing values were identified in the Description and CustomerID columns. CustomerID contains the majority of missing records and requires further attention before customer-level analysis can be performed.

---

### Missing Value Percentage Analysis

```python
(df.isnull().sum()/len(df))*100
```

Approximately 25.16% of records contain missing CustomerID values, while only 0.27% of records have missing product descriptions. Customer identifiers represent the primary data quality concern in the dataset.

---

### Missing Product Description Assessment

```python
df["Description"].isnull().sum()
```

A total of 1,454 transactions contain missing product descriptions, which may affect product-level analysis and reporting.

---

### Missing Product Description Handling

```python
df = df.dropna(subset=["Description"])
```

Records with missing product descriptions were removed to maintain the quality and reliability of product-based analysis.

---

### Missing Customer Identifier Assessment

```python
df["CustomerID"].isnull().sum()
```

A total of 133,583 transactions do not contain customer identifiers, making customer-level analysis and segmentation impossible for these records.

---

### Missing Customer Identifier Handling

```python
df = df.dropna(subset=["CustomerID"])
```

Transactions without customer identifiers were removed to ensure that all remaining records can be associated with a valid customer for CLV analysis and customer segmentation.

---

### Invalid Quantity Assessment

```python
(df["Quantity"] <= 0).sum()
```

A total of 8,872 transactions contain non-positive quantities, which likely represent cancelled orders, product returns, or invalid transaction records.

---

### Invalid Quantity Handling

```python
df = df[df["Quantity"] > 0]
```

Transactions with non-positive quantities were removed to ensure that only valid sales transactions remain in the dataset.

---

### Invalid Unit Price Assessment

```python
(df["UnitPrice"] <= 0).sum()
```

A small number of transactions (40 records) contain non-positive unit prices, which could negatively impact revenue calculations and business analysis.

---

### Invalid Unit Price Handling

```python
df = df[df["UnitPrice"] > 0]
```

Transactions with non-positive unit prices were removed to maintain pricing consistency and support accurate revenue calculations.

---

### Missing Values Validation

```python
df.isnull().sum()
```

All missing values have been successfully addressed. The cleaned dataset no longer contains null values in any column.

---

### Duplicate Records Validation

```python
df.duplicated().sum()
```

No duplicate records remain in the dataset, confirming that duplicate removal was successfully completed.

---

### Transaction Quality Validation

```python
(df["Quantity"] <= 0).sum(),
(df["UnitPrice"] <= 0).sum()
```

The dataset no longer contains transactions with non-positive quantities or non-positive unit prices, indicating successful transaction quality validation.

---

### Final Dataset Dimensions

```python
df.shape
```

The final cleaned dataset contains 392,692 valid transaction records across 8 columns. The dataset is now suitable for feature engineering, exploratory analysis, customer analytics, and segmentation tasks.

---

### Export Cleaned Dataset

```python
df.to_csv(
    "../outputs/cleaned_online_retail.csv",
    index=False
)

print("Cleaned dataset saved successfully.")
```

The cleaned dataset was successfully exported and saved for use in subsequent phases of the project, ensuring reproducibility and workflow continuity.
