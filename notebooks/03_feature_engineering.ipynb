# Retail Sales Data Analysis & Customer Segmentation

## Level 3: Feature Engineering

Feature Engineering is the process of creating new variables from existing data to improve analytical capabilities and generate deeper business insights.

### Objective

The objective of this phase is to transform cleaned transactional data into meaningful business-oriented features that enhance analytical capabilities.

These engineered features will support revenue analysis, trend identification, customer behavior analysis, customer segmentation, Customer Lifetime Value (CLV) evaluation, and business decision-making throughout the subsequent stages of the project.

### Revenue Feature Creation

```python
df["Revenue"] = df["Quantity"] * df["UnitPrice"]
```

A Revenue feature is successfully created by multiplying Quantity and UnitPrice.

This feature represents the monetary value generated by each transaction and will serve as the primary metric for sales performance analysis, customer value evaluation, Pareto analysis, RFM modeling, customer segmentation, and Customer Lifetime Value (CLV) assessment.

---

### Revenue Feature Validation

```python
df[["Quantity","UnitPrice","Revenue"]].head()
```

The Revenue feature is validated successfully. Sample records confirm that transaction revenue is accurately calculated as the product of Quantity and Unit Price.

The feature is ready for subsequent analytical and business intelligence tasks.

---

### Year Feature Creation

```python
df["Year"] = df["InvoiceDate"].dt.year
```

A Year feature is successfully extracted from the transaction date.

This feature supports time-based analysis and allows business performance comparisons across different years.

---

### Month Feature Creation

```python
df["Month"] = df["InvoiceDate"].dt.month
```

A Month feature is created successfully to support monthly sales analysis and seasonality identification.

This feature will help uncover patterns in customer purchasing behavior throughout the year.

---

### Month Name Feature Creation

```python
df["MonthName"] = df["InvoiceDate"].dt.month_name()
```

The MonthName feature is generated successfully to provide business-friendly month labels.

This improves the readability of reports, visualizations, and business presentations.

---

### Quarter Feature Creation

```python
df["Quarter"] = df["InvoiceDate"].dt.quarter
```

A Quarter feature is created successfully to support quarterly business performance analysis.

This feature enables revenue and sales comparisons across different quarters of the year.

---

### Day Of Week Feature Creation

```python
df["DayOfWeek"] = df["InvoiceDate"].dt.day_name()
```

The DayOfWeek feature is generated successfully to identify customer purchasing patterns across weekdays.

This feature will help determine which days contribute most to business activity and revenue generation.

---

### Invoice Month-Year Feature Creation

```python
df["InvoiceMonthYear"] = (
    df["InvoiceDate"]
      .dt.to_period("M")
      .astype(str)
)
```

The InvoiceMonthYear feature is successfully created by combining month and year information.

This feature simplifies monthly trend analysis and supports chronological reporting of sales performance.

---

### Weekend Indicator Creation

```python
df["IsWeekend"] = (
    df["InvoiceDate"]
      .dt.dayofweek >= 5
)
```

A weekend indicator feature is successfully created to distinguish between weekday and weekend transactions.

This feature will support analysis of customer purchasing behavior across different periods of the week.
